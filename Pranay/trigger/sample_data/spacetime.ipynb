```python
title = 'auto' # None  'auto'  'TITLE'

xlabel = 'auto' # 'auto'
ylabel = 'auto'  # 'auto'

xlim={} #{'xmin':1, 'xmax':50}
ylim={} #{'ymin':1, 'ymax':50}
colorRange = dict(vmin=0, vmax=1) # dict(vmin=0, vmax=1)


vertical_on_x = True # False  True
figsize = (10,6)

output = 'show' # 'show' 'png' 'pdf' 'eps'
```

```python
def dataframe(filename,xlabel,ylabel,vertical_on_x):
    data = pd.read_csv(filename, sep='\t', index_col=0)
    data.columns = data.columns.astype('int')
    data.index.name, data.columns.name = data.index.name.split('\\')
    if(vertical_on_x): data = data.transpose()
    if(xlabel!='auto'):data.columns.name=xlabel
    if(ylabel!='auto'):data.index.name=ylabel
    return data
```

```python
df = dataframe(filename,xlabel,ylabel, vertical_on_x)
```

```python
def get_title(filename,title):
    if(title=='auto'):
        title = ' '.join( filename[:-4].split('_') )
    return title
get_title(filename,title)
```

```python
plt.figure(figsize=figsize)
plt.title( get_title(filename,title) )
plt.xlim(**xlim)
plt.ylim(**ylim)

plt.imshow(df.values, aspect='auto', origin='lower', interpolation = 'none',
           extent=(df.columns.min(),df.columns.max(),
                   df.index.min(),df.index.max()), **colorRange )
plt.colorbar()
plt.tight_layout()
if(output=='show'):
    plt.show()
```
